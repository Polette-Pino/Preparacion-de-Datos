```python
# 4. Agrupe por state/region y ages, para luego agregar usando promedio, mínimo, máximo y conta

# Agrupando por state y ages, y luego calculando la media, mínimo, máximo y conteo de la población
poblacion_df.groupby(['state/region', 'ages'])['population'].agg(['mean', 'min', 'max', 'count'])
```

```python
# Agrupar por state/region y sumar la población
poblacion_suma = poblacion_df.groupby("state/region")["population"].sum().reset_index()

# Mostrar el resultado
poblacion_suma.head(10)
```

```python
# cambiar nombre de la columna state/region por abbreviation

poblacion_df2 = poblacion_suma.rename(columns={'state/region': 'abbreviation'})
poblacion_df2.head(5)
```

```python
# 5. Forme un nuevo dataset llamado df que tenga la información de los 3 sets importados. Use las llaves necesarias.

df = pd.merge(abreviaturas_df, areas_df, on='state', how='outer')
df = pd.merge(df, poblacion_df2, on='abbreviation', how='outer')

df.head(10)

# 6. Calcule la población por área para cada estado. 

# 7. Finalmente, cree un dataframe que posea: El estado, el área, la 
# población, la población por área y como índice la abreviación del 
# nombre del estado.
```
