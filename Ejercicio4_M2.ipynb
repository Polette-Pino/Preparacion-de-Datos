```python
# e) Mostrar por pantalla los nombres de las personas que iban en primera clase ordenadas alfabéticamente

# cerar una mascara de la columna Pclass para clase "1"
mascara_Pclass = titanic_df.Pclass == 1
#ordenar los de primera clase en orden alfabetico
titanic_df.loc[mascara_Pclass].sort_values("Name")
```

```python
# f) Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.

titanic_df["Survived"].value_counts(normalize=True) * 100
```

```python
# g) Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.

# calcular el porcentaje de personas que sobrevivieron en cada clase
sobrevivientes_por_clase = titanic_df.groupby('Pclass')['Survived'].mean() * 100
sobrevivientes_por_clase
```

```python
# normalizar los valores anteriores para que sumen 100%
normalizar_sobrev_por_clase = sobrevivientes_por_clase / sobrevivientes_por_clase.sum() * 100
normalizar_sobrev_por_clase
```

```python
# h) Eliminar del DataFrame los pasajeros con edad desconocida.

titanic_df.dropna(subset=['Age'])
```

```python
# i) Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.

titanic_df.groupby(['Pclass', 'Sex'])['Age'].mean().unstack()['female']
```

```python
# j) Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.

titanic_df['Young'] = titanic_df['Age'] < 18
titanic_df.head(10)
```

```python
# k) Mostrar por pantalla el porcentaje de menores y mayores de edad que sobrevivieron en cada clase

titanic_df.groupby(['Pclass', 'Young'])['Survived'].value_counts(normalize = True) * 100
```
